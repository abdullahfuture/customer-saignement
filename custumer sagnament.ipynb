## Clustring- Univariate , Bivariate ,Multivariate

```python
df.columns
```

```python
clustring1 = KMeans (n_clusters=3)
```

```python
clustring1.fit(df[['Annual Income (k$)']])
```

```python
clustring1.labels_
```

```python
df['income cluster']=clustring1.labels_
```

```python
df.head()
```

```python
df['income cluster'].value_counts()
```

```python
inertia_scores = [] 
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)  
    kmeans.fit(df[['Annual Income (k$)']])
    inertia_scores.append(kmeans.inertia_) 
```

```python
plt.plot(range(1, 11),inertia_scores)
```

```python
df.columns
```

```python
df.groupby(['income cluster'])[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()
```

Univariate

```python
clustring2 = KMeans(n_clusters=5)
clustring2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
df['spending and annual income cluster '] =clustring2.labels_
df.head()
```

```python
inertia_scores2 = [] 
for i in range(1, 11):
    kmeans2 = KMeans(n_clusters=i)  
    kmeans2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
    inertia_scores2.append(kmeans2.inertia_) 
plt.plot(range(1,11),inertia_scores2)
```

```python
centers = pd.DataFrame(clustring2.cluster_centers_)
centers.columns = ['Annual Income (k$)', 'Spending Score (1-100)'] 
```

```python

plt.figure(figsize=(10,8))
plt.scatter(x=centers['Annual Income (k$)'], y=centers['Spending Score (1-100)'], s=100, c='black', marker='*', label='Cluster Centers')
sns.scatterplot(data=df,x='Annual Income (k$)',y='Spending Score (1-100)',hue='spending and annual income cluster ',palette='tab10')
```
